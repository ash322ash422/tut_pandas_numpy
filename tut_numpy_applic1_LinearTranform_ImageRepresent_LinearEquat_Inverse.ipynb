# Linear Transformation

```python
# Lets do scaling of points
import numpy as np
import matplotlib.pyplot as plt
l = [
    [1, 1],
    [1, 2],
    [1, 3],
    [1.5, 3],
    [2, 3],
    [3, 3]
]
points = np.array(l)

# Scaling matrix (scale x by 2, y by 0.5)
scale_matrix = np.array([
    [2, 0],
    [0, 0.5]
])

# Apply the transformation
scaled_points = points @ scale_matrix.T  # (or np.dot(points, scale_matrix.T))

# Plot
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], color='blue', label='Original Points')
plt.scatter(scaled_points[:,0], scaled_points[:,1], color='red', label='Scaled Points')

# (OPTIONAL) Draw lines connecting original and transformed points
for i in range(len(points)):
    plt.plot([points[i,0], scaled_points[i,0]], [points[i,1], scaled_points[i,1]], 'k--', linewidth=0.8)

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Scaling of Points using Linear Transformation')
plt.grid(True)
plt.axis('equal')

plt.show()
```

```python
# Now lets do rotatin of points
import numpy as np
import matplotlib.pyplot as plt

l = [
    [1, 1],
    [1.5, 1.5],
    [2, 2],
    [3, 3],
    [4, 4],
    
]
points = np.array(l)


theta = np.pi / 2  # # Rotation angle: 90 degrees
# theta = np.pi / 4  # # Rotation angle: 45 degrees

# Rotation matrix
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Apply rotation
rotated_points = points @ rotation_matrix.T  # or np.dot(points, rotation_matrix.T)

# Plot
plt.figure(figsize=(6,6))
plt.scatter(points[:,0], points[:,1], color='blue', label='Original Points')
plt.scatter(rotated_points[:,0], rotated_points[:,1], color='green', label='Rotated Points')

# (OPTIONAL) Connect original and rotated points
for i in range(len(points)):
    plt.plot([points[i,0], rotated_points[i,0]], [points[i,1], rotated_points[i,1]], 'k--', linewidth=0.8)

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Rotation of 4 Points using Linear Transformation')
plt.grid(True)
plt.axis('equal')

plt.show()
```

```python
# Now lets do scaling and rotation
import numpy as np
import matplotlib.pyplot as plt

# Original triangle points
A = np.array([0, 0])
B = np.array([1, 0])
C = np.array([0, 1])

# Combine points into matrix
points = np.array([A, B, C])

# Scaling matrix
scale_matrix = np.array([
    [2, 0],
    [0, 3]
])

# Apply scaling
scaled_points = points @ scale_matrix.T

# Rotation matrix (45 degrees)
theta = np.pi / 4  # 45 degrees
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Apply rotation
rotated_scaled_points = scaled_points @ rotation_matrix.T

plt.figure(figsize=(6,6)) # Plotting

# Plot original
plt.fill(points[:,0], points[:,1], color='blue', alpha=0.3, label='Original Triangle')

# Plot scaled
plt.fill(scaled_points[:,0], scaled_points[:,1], color='green', alpha=0.3, label='Scaled Triangle')

# Plot rotated-scaled
plt.fill(rotated_scaled_points[:,0], rotated_scaled_points[:,1], color='red', alpha=0.3, label='Rotated + Scaled Triangle')

plt.legend()
plt.title('Triangle: Scaling and Rotation')
plt.grid(True)
plt.axis('equal')

plt.show()
```

# Image representation

```python
##  Image Representation (Grayscale Simulation)
import numpy as np
import matplotlib.pyplot as plt

l = [
    [255, 0, 156],
    [0, 0, 0],
    [0, 0, 255]
]
image = np.array(l) # Create a 3X3 image filled with 0 (black)

# print(f"image.shape: {image.shape}")
print(f"image: \n{image}")

plt.imshow(image, cmap='gray')

plt.title("X Pattern on Grayscale Image")
plt.colorbar()

plt.show()
```

```python
##  Draw X: Image Representation (Grayscale Simulation)

import numpy as np
import matplotlib.pyplot as plt

# Create a 10x10 image filled with 0 (black)
image = np.zeros((10, 10), dtype=np.uint8)

# Draw an "X" by setting both diagonals to 255 (white)
for i in range(10):
    image[i, i] = 255            # Main diagonal
    image[i, 9 - i] = 255        # Anti-diagonal

print(f"image.shape: {image.shape}")
print(f"image: \n{image}")

# Show the image
plt.imshow(image, cmap='gray')
plt.title("X Pattern on Grayscale Image")
plt.colorbar()
plt.show()
```

```python
# Lets apply filter convolution on a simple image (without padding)
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt

# 4x4 image
image = np.array([
    [100, 100, 100, 100],
    [100,  50,  50, 100],
    [100,  50, 255, 100],
    [100, 100, 100, 100],
])

# 2x2 averaging filter
filter = np.ones((2, 2)) / 4 # all values are 0.25

blurred_image = convolve2d(image, filter, mode='valid') #  Apply convolution without padding
# blurred_image = convolve2d(image, filter, mode='same', boundary='fill', fillvalue=0) #  Apply convolution with padding

# Show original and blurred image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Filtered (2x2)")
plt.imshow(blurred_image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.show()

# Also print the arrays
print("Original Image:")
print(image)

print("\nFiltered Image:")
print(np.round(blurred_image, 2))
```

```python
# Lets say we have two small color image patches, and we want to apply elementwise 
# operations such as brightness adjustment, image blending, and difference detection.

import numpy as np
import matplotlib.pyplot as plt

# Define two 2x3 RGB image patches
img1 = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
                 [[255, 255, 0], [0, 255, 255], [255, 0, 255]]], dtype=np.uint8)

img2 = np.array([[[100, 100, 100], [50, 50, 50], [150, 150, 150]],
                 [[20, 20, 20], [200, 200, 200], [10, 10, 10]]], dtype=np.uint8)

# Perform pixel-wise operations
added = np.clip(img1 + img2, 0, 255).astype(np.uint8)
brightened = np.clip(img1 + 20, 0, 255).astype(np.uint8)
subtracted = np.clip(img1.astype(int) - img2.astype(int), 0, 255).astype(np.uint8)

# Plotting
fig, axs = plt.subplots(2, 3, figsize=(10, 5))
axs[0, 0].imshow(img1)
axs[0, 0].set_title("Image 1")
axs[0, 0].axis('off')

axs[0, 1].imshow(img2)
axs[0, 1].set_title("Image 2")
axs[0, 1].axis('off')

axs[0, 2].imshow(added)
axs[0, 2].set_title("Added")
axs[0, 2].axis('off')

axs[1, 0].imshow(brightened)
axs[1, 0].set_title("Brightened (img1 + 20)")
axs[1, 0].axis('off')

axs[1, 1].imshow(subtracted)
axs[1, 1].set_title("Subtracted (img1 - img2)")
axs[1, 1].axis('off')

# Hide last subplot
axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

```python
## 4. Image Processing: Brightness Adjustment

import numpy as np
import matplotlib.pyplot as plt

# Simulate a grayscale image (values between 0 and 255)
image = np.random.randint(0, 256, (50, 50))
print(f"image= \n{image}")

# Brighten image by 50 (clip to max 255)
bright_image = np.clip(image + 50, 0, 255)
print(f"bright_image= \n{bright_image}")

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Brightened")
plt.imshow(bright_image, cmap='gray')

plt.show()
```

## 5. Solving Systems of Linear Equations

**Example:** Solve the following equations:

[ 2x + y = 5

x + 3y = 10 ]

```python
import numpy as np

# Coefficient matrix
l = [[2, 1],
    [1, 3]
]
A = np.array(l)

# Right-hand side vector
b = np.array([5, 10])

# Solve the system
solution = np.linalg.solve(A, b)
print("Solution [x, y]:", solution)
```

# inverse of matrix

```python
# inverse of matrix
import numpy as np

l = [[1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
]
A = np.array(l)

# Compute its inverse
A_inv = np.linalg.inv(A)

# Print the result
print("Inverse of matrix A:", A_inv)
```

```python
# now lets verify:
res = A @ A_inv
print(res)

res = A_inv @ A
print(res)
```
