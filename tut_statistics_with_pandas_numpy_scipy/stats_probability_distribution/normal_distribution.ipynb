```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Style1: Histogram

```python
# step1: 1 Histogram

np.random.seed(42)  # for reproducibility
mu = 50     # mean
sigma = 6  # standard deviation
n= 100    # 10, 50, 100, 500, 1000 160, ,   number of data points
    
data = np.random.normal(mu, sigma, n)

plt.style.use('dark_background')  # Black background
plt.figure(figsize=(10, 6))

plt.hist(data, bins=30, density=True, color='orange', alpha=0.6, edgecolor='white')

# Theoretical normal curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
y = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)
plt.plot(x, y, color='#FF1493', linewidth=2.5, label='Normal Curve (=50, =10)')

# Styling
plt.title("Normally Distributed Data", fontsize=16, color='white', pad=15)
plt.xlabel("Value", fontsize=12, color='white')
plt.ylabel("Density", fontsize=12, color='white')
plt.grid(alpha=0.2)
plt.legend(facecolor='black', edgecolor='white', fontsize=10)
plt.tight_layout()

plt.show()
```

```python

# Step 1: Setup
np.random.seed(42)
mu = 50
sigma = 6
sizes = [10, 50, 100, 500, 1000, 10000]

# Seaborn + dark background style
plt.style.use('dark_background')
sns.set(style='whitegrid', rc={'axes.facecolor':'black', 'figure.facecolor':'black'})

# Step 2: Create subplots (2 rows  3 columns)
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

# Step 3: Plot each distribution with dynamic bin width
for i, n in enumerate(sizes):
    data = np.random.normal(mu, sigma, n)
    
    # Dynamic bin calculation (Scotts or FreedmanDiaconis can also be used)
    # Here we simply make bin width proportional to sample size
    bin_width = sigma * 4 / np.sqrt(n)     # narrower bins as n grows
    bins = int((max(data) - min(data)) / bin_width)
    bins = max(5, bins)                    # ensure minimum bins
    
    sns.histplot(
        data, bins=bins, 
        # kde=True,
        color='orange', edgecolor='white', alpha=0.6, ax=axes[i]
    )
    
    # Labels and formatting
    axes[i].set_title(f"n = {n}   (bins = {bins})", fontsize=13, color='white', pad=10)
    axes[i].set_xlabel("Value", fontsize=11, color='white')
    axes[i].set_ylabel("Density", fontsize=11, color='white')
    axes[i].grid(alpha=0.2)
    axes[i].tick_params(colors='white')

# Step 4: Adjust layout
plt.suptitle("Effect of Sample Size on Normal Distribution (Dynamic Bin Width)", 
             fontsize=18, color='white', y=1.03)
plt.tight_layout()
plt.show()
```

### Style2 Stem plot

```python
# Stem plot

# Plot with dark background
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))

# Generate Normally Distributed Data
np.random.seed(42)
mu    = 50
sigma = 6
n     = 2000
data = np.random.normal(mu, sigma, n)

# Compute histogram data (for stem)
counts, bin_edges = np.histogram(data, density=True)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# Stem plot
(markerline, stemlines, baseline) = plt.stem(
    bin_centers, counts, linefmt='orange', markerfmt='o', basefmt=' ',
)
plt.setp(markerline, markersize=6, markerfacecolor='cyan', markeredgecolor='white')
plt.setp(stemlines, linewidth=2)

# (OPTIONA) Theoretical normal curve
# x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200)
# y = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)
# plt.plot(x, y, color='#FF1493', linewidth=2.5, label='Normal Curve (=50, =10)')

# Styling
plt.title("Normally Distributed Data (Stem Plot)", fontsize=16, color='white', pad=15)
plt.xlabel("Value", fontsize=12, color='white')
plt.ylabel("Density", fontsize=12, color='white')
plt.grid(alpha=0.2)
plt.legend(facecolor='black', edgecolor='white', fontsize=10)
plt.tight_layout()

# Show
plt.show()
```
