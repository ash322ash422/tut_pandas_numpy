# Positive or Right Skewed data
Examples of right skewedness :

Income Distribution: In many countries, a large portion of the population earns moderate incomes, while a smaller number of individuals earn extremely high incomes, creating a right-skewed distribution. 

Test Scores: In education, most students might score around the average, but a few exceptional students may achieve very high scores, resulting in a right-skewed distribution of test scores. 


```python
# Positive or Right Skewed data
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gamma, mode, skew

# Set seed for reproducibility
np.random.seed(42)

# Generate positively skewed data (Gamma distribution is more realistic)
data = gamma.rvs(a=2, scale=2, size=1000) + 2  # a=shape, scale=scale
data = np.round(data,1)
print(data[:20])


# Calculate statistics
data_mean = np.mean(data)
data_median = np.median(data)
data_mode = mode(data, keepdims=True)[0][0]
data_skewness = skew(data)

# Print stats
print(f"Mean    : {data_mean:.2f}")
print(f"Median  : {data_median:.2f}")
print(f"Mode    : {data_mode:.2f}")
print(f"Skewness: {data_skewness:.2f}")

# Plot histogram and KDE
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='skyblue', edgecolor='black')
plt.axvline(data_mean, color='red', linestyle='--', label=f"Mean: {data_mean:.2f}")
plt.axvline(data_median, color='green', linestyle='--', label=f"Median: {data_median:.2f}")
plt.axvline(data_mode, color='orange', linestyle='--', label=f"Mode: {data_mode:.2f}")
plt.title('Realistic Positively Skewed Data (Gamma Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

# Negative or Left Skewed data
An example of a real life variable that has a skewed left distribution is age of death from natural causes (heart disease, cancer, etc.). Most such deaths happen at older ages, with fewer cases happening at younger ages.

```python
# Negative or Left Skewed data
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gamma, mode, skew

# Set seed for reproducibility
np.random.seed(42)

# Generate negatively skewed data by reflecting a positively skewed Gamma distribution
positive_data = gamma.rvs(a=2, scale=2, size=1000)
data = -positive_data + np.max(positive_data) + 1  # Reflect and shift to keep values positive

# Scale to represent real-life age of death (approx. 4 to 100)
min_age = 4
max_age = 100
scaled_data = (data - np.min(data)) / (np.max(data) - np.min(data))  # scale to 0-1
data = scaled_data * (max_age - min_age) + min_age  # scale to 4100

data = np.round(data,0)
print(data[:20])


# Calculate statistics
data_mean = np.mean(data)
data_median = np.median(data)
data_mode = mode(data, keepdims=True)[0][0]
data_skewness = skew(data)

# Print stats
print(f"Mean    : {data_mean:.2f}")
print(f"Median  : {data_median:.2f}")
print(f"Mode    : {data_mode:.2f}")
print(f"Skewness: {data_skewness:.2f}")

# Plot histogram and KDE
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='lightcoral', edgecolor='black')
plt.axvline(data_mean, color='red', linestyle='--', label=f"Mean: {data_mean:.2f}")
plt.axvline(data_median, color='green', linestyle='--', label=f"Median: {data_median:.2f}")
plt.axvline(data_mode, color='blue', linestyle='--', label=f"Mode: {data_mode:.2f}")
plt.title('Realistic Negatively Skewed Data (Reflected Gamma Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

# Zero skewed data
Example: Height distribution


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, mode, skew

# Set seed for reproducibility
np.random.seed(42)

# Generate symmetric data using the normal distribution
data = np.random.normal(loc=100, scale=15, size=1000) + 20  # Mean=100, Std=15
data = np.round(data,0)
print(data[:20])

# Calculate statistics
data_mean = np.mean(data)
data_median = np.median(data)
data_mode = mode(data, keepdims=True)[0][0]
data_skewness = skew(data)

# Print stats
print(f"Mean    : {data_mean:.2f}")
print(f"Median  : {data_median:.2f}")
print(f"Mode    : {data_mode:.2f}")
print(f"Skewness: {data_skewness:.2f}")

# Plot histogram and KDE
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, bins=30, color='lightblue', edgecolor='black')
plt.axvline(data_mean, color='red', linestyle='--', label=f"Mean: {data_mean:.2f}")
plt.axvline(data_median, color='green', linestyle='--', label=f"Median: {data_median:.2f}")
plt.axvline(data_mode, color='orange', linestyle='--', label=f"Mode: {data_mode:.2f}")
plt.title('Zero Skewed Data (Normal Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
