# Boxplot

### Boxplot shows:
- Minimum (excluding outliers): The smallest data point still within 1.5 IQR of the lower quartile.
- First Quartile (Q1): 25% of the data fall below this value.
- Median (Q2): The middle value (50% of the data fall below this).
- Third Quartile (Q3): 75% of the data fall below this value.
- Maximum (excluding outliers): The largest data point still within 1.5 IQR of the upper quartile.

### Box Components:
- Box: Spans from Q1 to Q3 (the interquartile range, or IQR). It shows the middle 50% of the data.
- Line inside the box: The median.
- Whiskers: Extend from the box to the minimum and maximum values within 1.5 * IQR from Q1 and Q3.
- Dots (Outliers): Individual points that fall outside the whiskers (i.e., more than 1.5 * IQR from the quartiles).


### What You Can Learn from a Boxplot:
- 1) Central tendency (via the median).
- 2) Spread of the data (via the IQR and whiskers).
- 3) Skewness:

     If the median is closer to Q1 → right-skewed.

     If the median is closer to Q3 → left-skewed.

- 4) Presence of outliers.
- 5) Comparison between groups (boxplots are great for side-by-side comparison across categories).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# First lets deterine all the 5 point key summaries:
data = np.array([5, -2, 20, 13, 4, -20, 15, 8, 10, 12, 6, 40, 45, 1])
print("data        :", list(data))
print("data(Sorted):", list(np.sort(data)))

min_val = np.min(data)
Q1 = np.percentile(data, 25)
median = np.median(data) # Q2 is median
Q3 = np.percentile(data, 75)
max_val = np.max(data)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1
lower_bound = (Q1 - 1.5 * IQR)
upper_bound = (Q3 + 1.5 * IQR)

# Identify outliers (any data points outside 1.5 * IQR above Q3 or below Q1)
outliers = data[(data < lower_bound) | (data > upper_bound)]

min_whisker = np.min(data[data >= lower_bound])
max_whisker = np.max(data[data <= upper_bound])


# Print out the values
print(f"Stats:")
print(f"  min_val: {min_val:.2f}")
print(f"  lower_bound: {lower_bound:.2f}")
print(f"  min_whisker: {min_whisker}")
print(f"  Q1: {Q1:.2f}")
print(f"  Q2: Median: {median:.2f}")
print(f"  Q3: {Q3:.2f}")
print(f"  max_whisker: {max_whisker}")
print(f"  upper_bound: {upper_bound:.2f}")
print(f"  max_val: {max_val:.2f}")
print(f"  Outliers: \n{outliers}")
print()
```

```python
# Vertical boxplot


# Dark background style
plt.style.use('dark_background')

# Create figure
plt.figure(figsize=(8, 5))

# Define colors
box_face      = '#FFD700'        # gold box fill
edge_color    = '#FFF8DC'      # bright cream edges
median_color  = 'green'    # bright yellow median
outlier_color = '#FFCC00'   # strong yellow-orange outliers
whisker_color = '#FFEE88'   # bright whiskers

# Create boxplot with enhanced styles
sns.boxplot(
    data=data,
    width=0.3,
    boxprops=dict(facecolor=box_face, color=edge_color, linewidth=2.5),
    whiskerprops=dict(color=whisker_color, linewidth=2.2),
    capprops=dict(color=whisker_color, linewidth=2.2),
    medianprops=dict(color=median_color, linewidth=2.5),
    flierprops=dict(marker='o', markerfacecolor=outlier_color,
                    markeredgecolor='white', markersize=10, alpha=0.9)
)

# Titles and labels
plt.title('Boxplot: Data Distribution', color='gold', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Values', color='gold', fontsize=12)
plt.xticks([])
plt.yticks(color='gold')

# Subtle grid
plt.grid(True, color='#444444', linestyle='--', linewidth=0.6, alpha=0.7)

# Border styling
for spine in plt.gca().spines.values():
    spine.set_edgecolor('gold')
    spine.set_linewidth(1.2)

# Show plot
plt.tight_layout()
plt.show()
```

# Advanced applications

```python
# Create a function to calculate all boxplot statistics
def calculate_box_stats(data):
    # Calculate the box plot statistics
    min_val = np.min(data)
    Q1 = np.percentile(data, 25)
    median = np.median(data)
    Q3 = np.percentile(data, 75)
    max_val = np.max(data)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    lower_bound = (Q1 - 1.5 * IQR)
    upper_bound = (Q3 + 1.5 * IQR)
    
    # Identify outliers (any data points outside 1.5 * IQR above Q3 or below Q1)
    outliers = data[(data < lower_bound) | (data > upper_bound)]

    min_whisker = np.min(data[data >= lower_bound])
    max_whisker = np.max(data[data <= upper_bound])
    
    return min_val, lower_bound, min_whisker, Q1, median, Q3, max_whisker, upper_bound, max_val, outliers

# Print out the values
def print_box_stats(stats):
    min_val, lower_bound, min_whisker, Q1, median, Q3, max_whisker, upper_bound, max_val, outliers = stats
    print(f"Stats:")
    print(f"  min_val: {min_val:.2f}")
    print(f"  lower_bound: {lower_bound:.2f}")
    print(f"  min_whisker: {min_whisker}")
    print(f"  Q1: {Q1:.2f}")
    print(f"  Q2: Median: {median:.2f}")
    print(f"  Q3: {Q3:.2f}")
    print(f"  max_whisker: {max_whisker}")
    print(f"  upper_bound: {upper_bound:.2f}")
    print(f"  max_val: {max_val:.2f}")
    print(f"  Outliers: \n{outliers}")
    print()
```

```python
# Use dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8, 5))

# Coordinates for horizontal box
box_y = 0  # y-position for the box

# Draw the box (Q1 → Q3)
ax.add_patch(plt.Rectangle((Q1, box_y - 0.2), Q3 - Q1, 0.4,
                           facecolor='#FFD700', edgecolor='white', linewidth=2.5))

# Draw median line
ax.plot([median, median], [box_y - 0.2, box_y + 0.2],
        color='lime', linewidth=3)

# Draw whiskers
ax.plot([lower_whisker, Q1], [box_y, box_y], color='#FFEE88', linewidth=2)
ax.plot([Q3, upper_whisker], [box_y, box_y], color='#FFEE88', linewidth=2)

# Draw caps at whisker ends
ax.plot([lower_whisker, lower_whisker], [box_y - 0.1, box_y + 0.1], color='#FFEE88', linewidth=2)
ax.plot([upper_whisker, upper_whisker], [box_y - 0.1, box_y + 0.1], color='#FFEE88', linewidth=2)

# Plot outliers
ax.scatter(outliers, [box_y]*len(outliers),
           color='#FFAA00', s=100, edgecolor='white', zorder=3)

# Add text annotations
ax.text(Q1, box_y + 0.3, f"Q1={Q1:.2f}", color='gold', ha='center', fontsize=10)
ax.text(median, box_y + 0.4, f"Median={median:.2f}", color='lime', ha='center', fontsize=10)
ax.text(Q3, box_y + 0.3, f"Q3={Q3:.2f}", color='gold', ha='center', fontsize=10)

# Highlight bounds
ax.scatter([lower_bound, upper_bound], [box_y, box_y],
           color='white', s=80, label='Bounds (1.5×IQR)', zorder=4)
ax.text(lower_bound, box_y - 0.3, f"LB={lower_bound:.2f}", color='white', ha='center', fontsize=9)
ax.text(upper_bound, box_y - 0.3, f"UB={upper_bound:.2f}", color='white', ha='center', fontsize=9)

# Formatting
ax.set_yticks([])
ax.set_xlabel("Values", color='gold', fontsize=12)
ax.set_title("Manual Horizontal Boxplot (Using Computed Values)",
             color='gold', fontsize=14, fontweight='bold', pad=15)
ax.tick_params(axis='x', colors='gold')
ax.grid(True, color='#555', linestyle='--', alpha=0.5)
for spine in ax.spines.values():
    spine.set_edgecolor('gold')

plt.tight_layout()
plt.show()
```
