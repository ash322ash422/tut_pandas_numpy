## Cantilever Beam Deflection

A cantilever beam is like a diving board: fixed at one end, free at the other. When you stand on it, it bends downward. Engineers use simple formulas to predict how much it bends, so it’s safe and won’t break.

```
y = (F * x**2) * (3*L - x) / (6 * E * I)
```
You’re computing the deflection(y) at any point x along the beam’s length for a point load at the free end.

x: position along the beam.

![CANTILEVER](images/cantilever.png)

The formula calculates how much it bends at each x. The maximum bend is at the free end (x = L).

F = Force or load (Newtons)

L = Length of the beam (meters)

E = Young’s Modulus (material stiffness, Pascals)

I = Second Moment of Area (depends on beam cross-section, m⁴)


````python
# Now plot x-axis and y-axis for series of values: x is the distance along the beam;y is the vertical bend at x
import numpy as np
import matplotlib.pyplot as plt

# Beam deflection: y = (F*x**2)*(3*L - x)/(6*E*I)
F = 1000  # Force in Newtons
L = 2     # Length in meters
E = 2e11  # Young's modulus (Pa)
I = 1e-6  # Second moment of area (m^4)

x = np.linspace(0, L, 100)# generate 100 x points from 0 to L. 
x = np.round(x,4)
print(f"1st 10 values of x: {x[:10]}")

y = (F * x**2) * (3*L - x) / (6 * E * I)
print(f"1st 10 values of y: {y[:10]}")

plt.plot(x, y)

plt.xlabel('x: Length along beam (m)')
plt.ylabel('y: Deflection (m)')
plt.title('Cantilever Beam Deflection')

plt.show()
````

````python
# Show me the bend along the x-axis.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
F = 1000   # N
L = 2      # m
E = 2e11   # Pa
I = 1e-6   # m^4

# Position along beam
x = np.linspace(0, L, 100)# generate 100 points ranging from 0 to L
print(f"1st 10 x values: {x[:10]}")

# Deflection
y = (F * x**2) * (3*L - x) / (6 * E * I)
print(f"1st 10 y values: {y[:10]}")


# Plot the undeformed and deformed shapes
plt.figure(figsize=(10, 4))

# Original (undeformed) beam: a straight line
plt.plot(x, np.zeros_like(x), 'k--', label='Original Beam Axis')

# Deformed shape
plt.plot(x, -y, 'b-', label='Deflected Shape') # -y to display along negative y-axis

plt.xlabel('Length along beam (m)')
plt.ylabel('Deflection (m)')
plt.title('Cantilever Beam: Original vs. Deflected Shape')
plt.legend()
plt.grid(True)

plt.show()
````

````python
# OPTIONAL
# Interactive Cantilever Beam Deflection
# Would only work with widget is installed


import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to compute deflection
def beam_deflection(F, L, E):
    I = 1e-6  # m^4, assume fixed for this demo

    x = np.linspace(0, L, 200)
    y = (F * x**2) * (3*L - x) / (6 * E * I)

    plt.figure(figsize=(10, 4))
    plt.plot(x, -y * 1000, label='Deflection')  # scale for better display
    plt.plot(x, np.zeros_like(x), 'k--', label='Original Axis')
    plt.xlabel('Length along beam (m)')
    plt.ylabel('Deflection (mm)')
    plt.title(f'Cantilever Beam Deflection\nForce: {F} N, Length: {L} m, E: {E:.1e} Pa')
    plt.legend()
    plt.grid(True)
    plt.show()

# Make sliders for:
# Load (F), Length (L), and Young's Modulus (E)
interact(
    beam_deflection,
    F=FloatSlider(value=1000, min=100, max=5000, step=100, description='Force (N)'),
    L=FloatSlider(value=2, min=0.5, max=5, step=0.1, description='Length (m)'),
    E=FloatSlider(value=2e11, min=1e10, max=3e11, step=1e10, description='E (Pa)')
);
````
