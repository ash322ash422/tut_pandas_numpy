```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis, describe

# Set seed for reproducibility
np.random.seed(0)

# 1. Mesokurtic: Normal Distribution
data_mesokurtic = np.random.normal(0, 1, 1000)

# 2. Leptokurtic: t-distribution with low df (heavy tails)
data_leptokurtic = np.random.standard_t(df=3, size=1000)

# 3. Platykurtic: Uniform distribution (light tails)
data_platykurtic = np.random.uniform(-np.sqrt(3), np.sqrt(3), 1000)

# Store datasets and labels
datasets = [
    ('Mesokurtic (Normal)', data_mesokurtic),
    ('Leptokurtic (t-distribution)', data_leptokurtic),
    ('Platykurtic (Uniform)', data_platykurtic),
]

# Plot
plt.figure(figsize=(15, 5))
for i, (label, data) in enumerate(datasets, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data, kde=True, bins=30, color='skyblue')
    plt.title(f"{label}\nKurtosis = {kurtosis(data):.2f}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Show full statistics
for label, data in datasets:
    stats = describe(data)
    print(f"\n {label}")
    print(f"Mean: {stats.mean:.2f}")
    print(f"Std Dev: {np.sqrt(stats.variance):.2f}")
    print(f"Min: {stats.minmax[0]:.2f}, Max: {stats.minmax[1]:.2f}")
    print(f"Skewness: {stats.skewness:.2f}")
    print(f"Kurtosis (Excess): {stats.kurtosis:.2f}")
```
