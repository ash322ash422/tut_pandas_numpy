```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt

# Example: Fit y = a*x + b to noisy data
# I generated data using this
# x_data = np.linspace(0, 10, 20)
# x_data = np.round(x_data,2)
# y_data = 3 * x_data + 2 + np.random.normal(size=len(x_data))
# y_data = np.round(y_data,2)

x_data = [ 0, 0.53,  1.05,  1.58,  2.11,  2.63,  3.16,  3.68,  4.21,  4.74,  5.26,  5.79, \
  6.32,  6.84,  7.37,  7.89,  8.42,  8.95,  9.47, 10 ]
y_data= [ 1.42,  2.97,  5.34,  7.64,  7.13, 11.03, 11.08, 13.57, 16.36, 16.96, 15.99, 19.31, \
 20.21, 21.32, 23.78 ,25.77, 25.38, 29.42, 30.53, 30.88]

x_data = np.array(x_data)
y_data = np.array(y_data)

print(f"x_data:{x_data}")
print(f"y_data:{y_data}")



def model(x, a, b):
    return a * x + b

popt, pcov = curve_fit(model, x_data, y_data)
# popt is an array of the best-fit parameters for your model.
# In above case: popt = [a_opt, b_opt]
# These are the estimated values of a and b

# pcov is a 2D array (matrix) that estimates the covariance between the parameters.
print(f"popt = {popt}")
print(f"pcov = \n{pcov}")

plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, model(x_data, *popt), 'r-', label='Fitted line')

plt.legend()
plt.show()

print(f"Fitted parameters: a = {popt[0]:.2f}, b = {popt[1]:.2f}")
```
