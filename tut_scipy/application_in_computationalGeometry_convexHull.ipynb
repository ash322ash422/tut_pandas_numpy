```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull

# Step 1: Generate some random 2D points
np.random.seed(42)  # for reproducibility
points = np.random.rand(10, 2)*10  # 10 random points in 2D
print(f"points:\n{points}")

# Step 2: Compute the convex hull
hull = ConvexHull(points)
hull_indices = hull.vertices #The indices of the hull points
hull_points = points[hull_indices]# The actual coordinates of the hull points
print("Indices of hull points:", hull_indices)
print("Coordinates of hull points:\n", hull_points)
print("hull.simplices:\n", hull.simplices) # list of edges that form the convex hull.

# Step 3: Plot the points and the convex hull
plt.figure(figsize=(8, 6))
plt.plot(points[:, 0], points[:, 1], 'o', label='Points')

# Draw the convex hull edges
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k-')

plt.fill(points[hull.vertices, 0], points[hull.vertices, 1], 'lightblue', alpha=0.4, label='Convex Hull')
plt.title('Convex Hull of Random 2D Points')
plt.legend()
plt.show()
```
