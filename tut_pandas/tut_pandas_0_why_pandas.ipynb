# Pandas: An Introduction

## pandas
- Used for Exploratory Data Analysis
- used for importing/exporting data, data cleaning, creating/deleting columns, etc


### why pandas over SQL ?
- While SQL can perform aggregations and basic calculations, it is not designed for complex statistical analysis, machine learning, or advanced data visualization.
- SQL queries are typically executed as a whole, making iterative data exploration and transformation less intuitive compared to a programmatic approach.
- Pandas offers a rich set of functions for handling missing values, reshaping data, merging, grouping, and applying custom functions.
- **Integration with Python ecosystem** gives you rich set of tools
