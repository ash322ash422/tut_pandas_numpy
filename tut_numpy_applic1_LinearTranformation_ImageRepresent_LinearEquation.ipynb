## Linear Transformation
Original Triangle (Before Transformation): Let's take a triangle with these 2D points.

A = (0, 0)

B = (1, 0)

C = (0, 1)

This forms a right-angled triangle with area = 0.5

We’ll apply a 2×2 transformation matrix:
```
T = [[2, 1],
     [1, 3]]
```
This matrix will shear and stretch the triangle.

- When you apply a linear transformation (e.g., stretching, rotating, shearing) to a 2D geometric shape using a matrix (T), the absolute value of the determinant of that matrix tells you how much the shape's area or volume is scaled.

 **New Area = ∣det(T)∣ × Original Area**

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import det

# Original triangle points
A = np.array([0, 0])
B = np.array([1, 0])
C = np.array([0, 1])

triangle = np.array([A, B, C, A]).T  # shape: (2, 4) for plotting
print(f"triangle: \n{triangle}")

# Transformation matrix
T = np.array([[2, 1],
              [1, 3]])

# Apply transformation
transformed_triangle = T @ triangle

# Compute original and transformed area
original_area = 0.5  # For the triangle with base=1, height=1
det_T = abs(det(T))
new_area = det_T * original_area

# Plotting
plt.figure(figsize=(6,6))
plt.plot(triangle[0], triangle[1], 'b-', label='Original Triangle')
plt.plot(transformed_triangle[0], transformed_triangle[1], 'r-', label='Transformed Triangle')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.title(f"Area scaled by |det(T)| = {det_T:.2f}\nOriginal Area = {original_area}, New Area = {new_area:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
````

````python
##  Draw X: Image Representation (Grayscale Simulation)

import numpy as np
import matplotlib.pyplot as plt

# Create a 10x10 image filled with 0 (black)
image = np.zeros((10, 10), dtype=np.uint8)

# Draw an "X" by setting both diagonals to 255 (white)
for i in range(10):
    image[i, i] = 255            # Main diagonal
    image[i, 9 - i] = 255        # Anti-diagonal

print(f"image.shape: {image.shape}")
print(f"image: \n{image}")

# Show the image
plt.imshow(image, cmap='gray')
plt.title("X Pattern on Grayscale Image")
plt.colorbar()
plt.show()
````

````python
# Let’s say we have two small color image patches, and we want to apply elementwise 
# operations such as brightness adjustment, image blending, and difference detection.

import numpy as np
import matplotlib.pyplot as plt

# Define two 2x3 RGB image patches
img1 = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
                 [[255, 255, 0], [0, 255, 255], [255, 0, 255]]], dtype=np.uint8)

img2 = np.array([[[100, 100, 100], [50, 50, 50], [150, 150, 150]],
                 [[20, 20, 20], [200, 200, 200], [10, 10, 10]]], dtype=np.uint8)

# Perform pixel-wise operations
added = np.clip(img1 + img2, 0, 255).astype(np.uint8)
brightened = np.clip(img1 + 20, 0, 255).astype(np.uint8)
subtracted = np.clip(img1.astype(int) - img2.astype(int), 0, 255).astype(np.uint8)

# Plotting
fig, axs = plt.subplots(2, 3, figsize=(10, 5))
axs[0, 0].imshow(img1)
axs[0, 0].set_title("Image 1")
axs[0, 0].axis('off')

axs[0, 1].imshow(img2)
axs[0, 1].set_title("Image 2")
axs[0, 1].axis('off')

axs[0, 2].imshow(added)
axs[0, 2].set_title("Added")
axs[0, 2].axis('off')

axs[1, 0].imshow(brightened)
axs[1, 0].set_title("Brightened (img1 + 20)")
axs[1, 0].axis('off')

axs[1, 1].imshow(subtracted)
axs[1, 1].set_title("Subtracted (img1 - img2)")
axs[1, 1].axis('off')

# Hide last subplot
axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
````

````python
## 4. Image Processing: Brightness Adjustment

import numpy as np
import matplotlib.pyplot as plt

# Simulate a grayscale image (values between 0 and 255)
image = np.random.randint(0, 256, (50, 50))
print(f"image= \n{image}")

# Brighten image by 50 (clip to max 255)
bright_image = np.clip(image + 50, 0, 255)
print(f"bright_image= \n{bright_image}")

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Brightened")
plt.imshow(bright_image, cmap='gray')

plt.show()
````

## 5. Solving Systems of Linear Equations

**Example:** Solve the following equations:

[ 2x + y = 5

x + 3y = 10 ]

````python
import numpy as np

# Coefficient matrix
A = np.array([[2, 1], [1, 3]])

# Right-hand side vector
b = np.array([5, 10])

# Solve the system
solution = np.linalg.solve(A, b)
print("Solution [x, y]:", solution)
````
