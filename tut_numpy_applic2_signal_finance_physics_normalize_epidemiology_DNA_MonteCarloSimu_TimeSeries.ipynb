## 1. (OPTIONAL) Signal Processing: Smoothing a Noisy Signal

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate noisy signal
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)

# Moving average smoothing
window = 5
y_smooth = np.convolve(y, np.ones(window)/window, mode='valid')

plt.plot(x, y, label='Noisy Signal')
plt.plot(x[window-1:], y_smooth, label='Smoothed Signal', color='red')
plt.legend()
plt.show()
```

## 2. Financial Analysis: Compound Interest

```python
import numpy as np

# Parameters
principal = 1000  # Initial investment
rate = 0.05       # 5% interest
years = np.arange(1, 21)
print("Years:\n", years)

# Compound interest formula: A = P(1 + r)^t
amount = principal * (1 + rate) ** years
print("Investment value over years:\n", amount)
```

## 3. Physics Simulation: Free Fall under Gravity

```python
import numpy as np
import matplotlib.pyplot as plt

# Time steps
t = np.linspace(0, 10, 20)
print("Time steps:\n", t)

# Parameters
g = 9.81  # Acceleration due to gravity (m/s²)
v0 = 0    # Initial velocity
y0 = 100  # Initial height

# y = y0 + v0*t - 0.5*g*t^2
y = y0 + v0*t - 0.5*g*t**2
print("Height over time:\n", y)

plt.scatter(t, y)
plt.title("Object Falling Under Gravity")
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.grid()
plt.show()
```

## 5. Data Normalization (ML Preprocessing)

```python
import numpy as np

data = np.array([50, 20, 30, 80, 100], dtype=float)

# Min-Max Normalization: (x - min) / (max - min)
normalized = (data - np.min(data)) / (np.max(data) - np.min(data))
print("Normalized data:", normalized)
```

## 6. (OPTIONAL)Epidemiology: Simulating Infection Spread (Basic SIR Model)

```python
import numpy as np
import matplotlib.pyplot as plt

# Basic parameters
days = 160
S = np.zeros(days)  # Susceptible
I = np.zeros(days)  # Infected
R = np.zeros(days)  # Recovered

# Initial values
S[0], I[0], R[0] = 0.99, 0.01, 0.0
beta, gamma = 0.3, 0.1

# Simulation
for t in range(1, days):
    S[t] = S[t-1] - beta * S[t-1] * I[t-1]
    I[t] = I[t-1] + beta * S[t-1] * I[t-1] - gamma * I[t-1]
    R[t] = R[t-1] + gamma * I[t-1]

# Time array
time = np.arange(days)

# Line Plot
plt.figure(figsize=(10, 5))
# plt.plot(time, S, label='Susceptible', color='blue')
# plt.plot(time, I, label='Infected', color='red')
# plt.plot(time, R, label='Recovered', color='green')
plt.scatter(time, S, color='blue', s=10)
plt.scatter(time, I, color='red', s=10)
plt.scatter(time, R, color='green', s=10)
plt.title("SIR Epidemic Model with Line and Scatter Plot")
plt.xlabel("Days")
plt.ylabel("Population Fraction")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Random Simulations: Monte Carlo for Estimating Pi
# Image a square of side 2: [-1,1] X [-1,1]
# Probability that a random point is in the quarter circle of this square =  Area_circle / Area_square
# => (Number_of_points_in_quarter_circle) / (Total_number_of_points_in_square) =  (PI/4) / 1
# => PI = (Number_of_points_in_quarter_circle) / (Total_number_of_points_in_square) * 4

import numpy as np
np.random.seed(42)

# Number of random points
N = 1000 #1000000  # DO NOT MAKE IT 1000000000

# Generate random (x, y) pairs between -1 and 1
x = np.random.uniform(-1, 1, N) # generate N random numbers between -1 and +1.
y = np.random.uniform(-1, 1, N)
# print(f"x: {x}")

# Calculate distances from origin
distance_squared = x**2 + y**2
# print(f"distance_squared: {distance_squared}")

# Count how many points fall inside the unit circle
inside_circle = np.sum(distance_squared <= 1)
print(f"inside_circle: {inside_circle}")

# Estimate π
pi_estimate = 4 * inside_circle / N

print(f"Estimated π: {pi_estimate}")
```

```python
# (UNDERSTAND THIS) Time Series Analysis: Autocorrelation
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)  # Simulate a time series
autocorrelation = np.correlate(data, data, mode='full')
autocorrelation = autocorrelation[autocorrelation.size // 2:]

plt.plot(autocorrelation)
plt.title("Autocorrelation of Time Series")
plt.xlabel("Lag")
plt.show()
```
