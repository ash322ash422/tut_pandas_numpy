```python
import numpy as np
```

```python
# Create an array of numbers
arr = np.array([1, 2, 3, 4, 5])
print("Original Array:", arr)
```

```python
arr = np.array([1, 2, 9, 4, 25])
# np.sqrt() computes square root of each element
sqrt_arr = np.sqrt(arr)
print("Square Root:", sqrt_arr)
```

```python
arr = np.array([1, 2, 9, -4, 25])

# np.power() raises each element to the given power
power_arr = np.power(arr, 3)   # cube each element
print("Power (cube):", power_arr)
```

```python
arr = np.array([1, 2, 9, 4, 25])

# np.log() gives the natural logarithm (base e)
log_arr = np.log(arr)
print("Logarithm (base e):", log_arr)
```

```python
arr = np.array([1, 2, 9, -4, 25])

# np.exp() computes e^x for each element
exp_arr = np.exp(arr)
print("Exponential (e^x):", exp_arr)
```

```python
# combine
arr = np.array([1, 2, 9, -4, 25])

# np.exp() computes e^x for each element
result_arr = np.log(np.sqrt(np.exp(arr)))
print(result_arr)
```
