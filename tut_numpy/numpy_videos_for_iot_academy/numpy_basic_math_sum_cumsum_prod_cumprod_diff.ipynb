```python
import numpy as np
```

```python
# Create a simple integer array
arr = np.array([1.3, 2.45, -3, 4.7645, 5])
print("Original Array:", arr)
```

```python
# np.sum() returns the sum of all elements
total = np.sum(arr)
print("Sum of all elements:", total)
```

```python
arr = np.array([1.3, 2.45, -3, 4.7645, 5])

# np.cumsum() returns cumulative sum of elements
cumulative_sum = np.cumsum(arr)
print("Cumulative Sum:", cumulative_sum)
```

```python
arr = np.array([1.3, 2.45, -3, 4.7645, 5])

# np.prod() returns the product of all elements
product = np.prod(arr)
print("Product of all elements:", product)
```

```python
arr = np.array([1.3, 2.45, -3, 4.7645, 5])

# np.cumprod() returns cumulative product
cumulative_product = np.cumprod(arr)
print("Cumulative Product:", cumulative_product)
```

```python
arr = np.array([1.3, 2.45, -3, 4.7645, 5])

# np.diff() gives the difference between consecutive elements
difference = np.diff(arr)
print("Difference between consecutive elements:", difference)

# 1, 2    --> 2-1 =1
# 2, -3   --> -3 -2 = -5
# -3, 4   --> 4 - (-3) = 7
```

```python
arr = np.array([1, 2, -3, 4, 5])

# Stack results for visual comparison (note: diff has one less element)
result = {
    "Sum": total,
    "Cumulative Sum": cumulative_sum,
    "Product": product,
    "Cumulative Product": cumulative_product,
    "Difference": difference
}

for key, value in result.items():
    print(f"{key}: {value}")
```
