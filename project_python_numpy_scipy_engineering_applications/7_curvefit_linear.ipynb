## Linear Fit
Suppose an engineer measures the extension (mm) of a spring under different loads (N). He wants to determine if Hookes law holds:
(Hookeâ€™s Law says: **F = kx**, so force and extension should be linearly related)

Suppose we know that **extension (x) is proportional to force (F)**. So,

extension(x) = a * force + b

We do some experiment and measure F and x, and then using the data from this experiment , determine a and b


````python
import numpy as np
import matplotlib.pyplot as plt

# Suppose these are experimental measurements
force = np.array([0, 2, 4, 6, 8, 10])  # Force in Newton
extension = np.array([0, 1.1, 2.1, 3.0, 4.1, 5.1])  # Extension in mm

# Fit a line: degree 1 polynomial
coeffs = np.polyfit(force, extension, 1)

print("Slope and intercept:", coeffs)  # [a, b]
````

### result
So based on above data, we conclude that relation between x and F is following:
```
x = 0.5*F + 0.03
or
F = 2x - 0.06
```
Since 0.06 is small compared to the other term, we rewrite above: **F = 2x**
Hookes law holds here.

````python
# Now lets plot the above curve
# Make a polynomial function from coefficients
fit_fn = np.poly1d(coeffs)

# Predict extension for the same force values
predicted_extension = fit_fn(force)

# Plot
plt.scatter(force, extension, color='red', label='Data Points')
plt.plot(force, predicted_extension, color='blue', label='Fitted Line')

plt.xlabel('Force (N)')
plt.ylabel('Extension (mm)')
plt.title('Linear Fit: Hooke\'s Law')
plt.legend()
plt.show()
````
